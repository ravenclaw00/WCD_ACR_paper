```python
# Perform continuous wavelet transform with FFT-based algorithm
import pywt
import numpy as np
import matplotlib.pyplot as plt
for i in range(0,5):
    # Load data from file
    data = np.loadtxt('N:/atm_glomod/user/jomuel001/CMIP6_models/ICON/AREA.-90_90_89.7849_29.0866/EOF/PCslp_ICON_1985-2014_ollie_djfm_abg_aacrm21_remap1_cos_cov_svd')[:,i]
    print(data)

    wavelet = 'morl'  # Define wavelet type
    scales = np.linspace(1, 3630,3630)  # Define scales for analysis
    coefs, freqs = pywt.cwt(data, scales, wavelet, method='fft')
    fig, ax = plt.subplots(figsize=(8, 6))
    im = ax.contourf(scales, freqs*122, abs(coefs), cmap='coolwarm')
    ax.set_yscale('log')
    ax.set_ylim(0.1,10)
    ax.set_xlabel('Time')
    ax.set_ylabel('Frequency')
    plt.colorbar(im)
    #plt.show()
    plt.savefig('N:/atm_glomod/user/jomuel001/CMIP6_models/ICON/AREA.-90_90_89.7849_29.0866/PLOTS/WAVELET_slp_ICON_1985-2014_ollie_djfm_abg_aacrm21_remap1_cos_cov_unnorm'+str(i)+".png",dpi=300)
    plt.show()
```

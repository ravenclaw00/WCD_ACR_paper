```python
import xarray as xr
import os as os
import pandas as pd
from netCDF4 import Dataset as NetCDFFile 
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.basemap import Basemap
```

```python
path = r"N:/atm_glomod/user/jomuel001/CMIP6_models/ERA5/"
path_file = r"slp_hpa_ERA5_1979-2014.N_jjas_atrbg_aacrm21_remapbnds.nc"

fig = plt.figure(figsize=(8, 8))
map = Basemap(projection='ortho',lat_0=90, lon_0=360)
nc = NetCDFFile(path+path_file)
#print(nc)
nc

lat = nc.variables['lat'][:]
lon = nc.variables['lon'][:]
time = nc.variables['time'][:]
slp = nc.variables['MSL'][:] # mean sea level pressure

timestamp = 5
lons,lats = np.meshgrid(lon,lat)
x,y = map(lons,lats)
psl = map.contourf(x,y,slp[timestamp,:,:],cmap = 'bwr')
cb = map.colorbar(psl, "bottom", size="5%", pad="2%")
plt.title('sea level pressure')
cb.set_label('pressure (hPa)')
#print(np.size(time))
map.drawcoastlines()
map.drawcountries()

print(time[timestamp])
Jahr = int(str(time[timestamp])[:2])+1943
Monat = int(str(time[timestamp])[2])
Tag = int(str(time[timestamp])[4])
print('Datum : '+str(Tag)+'.'+str(Monat)+'.'+str(Jahr))
```

```python
Modell = 'ERA5'
AREA = "-90_90"
for Nummer_cluster in range(1,6,1):
    filename = r"KMEANSslp_hpa_"+Modell+r"_1979-2014_jjas_atrbg_aacrm21_remapbnds_unnorm_5tabcluster_10tabPC_dates"
    ordner = r"N:/atm_glomod/user/jomuel001/CMIP6_models/"+Modell+r"/AREA."+ str(AREA) + "_89.7849_29.0866/CLUSTER/"

    path = r"N:/atm_glomod/user/jomuel001/CMIP6_models/"+ Modell + "/"
    path_file = r"slp_hpa_ERA5_1979-2014.N_jjas_atrbg_aacrm21_remapbnds.nc"
    nc = NetCDFFile(path+path_file)

    #load txt and load variables
    cluster_assignment = np.loadtxt(ordner+filename)
    print(np.shape(cluster_assignment))
    lat = nc.variables['lat'][:]
    lon = nc.variables['lon'][:]
    time = nc.variables['time_bnds'][:]
    try:
        slp = nc.variables['MSL'][:]
    except: 
        slp = nc.variables['psl'][:]
    #print(slp)
    print(np.shape(slp))
    #assign data and cluster
    timestamps_1 = np.zeros((len(cluster_assignment[:,0]),3))
    timestamps_2 = np.zeros((len(cluster_assignment[:,0]),3))
    timestamps_3 = np.zeros((len(cluster_assignment[:,0]),3))
    timestamps_4 = np.zeros((len(cluster_assignment[:,0]),3))
    timestamps_5 = np.zeros((len(cluster_assignment[:,0]),3))
    
    boolean_1 = np.full((len(cluster_assignment), len(lat), len(lon)), False)
    boolean_2 = np.full((len(cluster_assignment), len(lat), len(lon)), False)
    boolean_3 = np.full((len(cluster_assignment), len(lat), len(lon)), False)
    boolean_4 = np.full((len(cluster_assignment), len(lat), len(lon)), False)
    boolean_5 = np.full((len(cluster_assignment), len(lat), len(lon)), False)

    #assighn psl values
    for  i in range(len(cluster_assignment[:,0])):
        read_cl = cluster_assignment[i,3]
        if read_cl == 1:
            timestamps_1[i,:] = cluster_assignment[i,:3]
            boolean_1[i,:,:] = True
        elif read_cl == 2:
            timestamps_2[i,:] = cluster_assignment[i,:3]
            boolean_2[i,:,:] = True
        elif read_cl == 3:
            timestamps_3[i,:] = cluster_assignment[i,:3]
            boolean_3[i,:,:] = True
        elif read_cl == 4:
            timestamps_4[i,:] = cluster_assignment[i,:3]
            boolean_4[i,:,:] = True
        elif read_cl == 5:
            timestamps_5[i,:] = cluster_assignment[i,:3]
            boolean_5[i,:,:] = True
    #create arrays with 




    #mean for cluster number 1  
    booleans = ["nix",boolean_1, boolean_2, boolean_3, boolean_4, boolean_5]

    fig = plt.figure(figsize=(8, 8))
    map = Basemap(projection='ortho',lat_0=90, lon_0=360)


    #slp_mean = np.zeros((80,320))
    slp_mean = np.ndarray.mean(slp[:,:,:], axis = 0, where = booleans[Nummer_cluster]) #mean of all days in ts 1

    lons,lats = np.meshgrid(lon,lat)
    x,y = map(lons,lats)
    psl = map.contourf(x,y,slp_mean[:,:],cmap = 'bwr',vmin =-np.max(np.abs(slp_mean[:,:])), vmax = np.max(np.abs(slp_mean[:,:])))
    cb = map.colorbar(psl, "bottom", size="5%", pad="2%")
    plt.title('sea level pressure, '+ str(Modell) + ", from AREA ATLANTIC -90-90" )
    cb.set_label('pressure (hPa)')
    #print(np.size(time))
    map.drawcoastlines()
    map.drawcountries()

    print("Clusternumber"+ str(Nummer_cluster) +  ", " + str(AREA) + str(Modell))
    print(str(AREA))
    plt.savefig(str(ordner) + "/PLOTS/slp_" + str(AREA) + "_1979-2014_10PC_" + "mean_Cl" + str(Nummer_cluster) + "assigment.png"  ,
                dpi=300,bbox_inches='tight')
```

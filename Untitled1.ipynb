```python
import pandas as pd

df = pd.read_excel(r'N:/atm_glomod/user/jomuel001/Dokumente/MasterThesis/figures/DatenSSPs.xlsx', engine='openpyxl', skiprows=None)
print(df)
years=df['Unnamed: 0']
SSP126=df['Unnamed: 1']
SSP245=df['Unnamed: 2']
SSP370=df['Unnamed: 3']
SSP585=df['Unnamed: 4']
```

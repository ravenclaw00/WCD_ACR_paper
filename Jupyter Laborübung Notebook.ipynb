_FORMELN_ <br>
Wir verwenden folgende Formel für die Schwingungsdauer eines Fadenpendels: <br>
$ T = 2\pi\cdot\surd \frac{l}{g} $ <br>
und stellen sie nach der Gravitationskonstante um: <br>
$g = \frac{4 \cdot {\pi}^2 \cdot l}{T^2}$

_MESSERGEBNISSE_ <br> 
Bei der Länge sind uns nur die Ableseungenauigkeit für das analoge Maßband bekannt: <br> $u_l=\frac{0,01cm}{2 \cdot \sqrt 6} = 2,04*10^{-3} cm$ <br>
Die Länge beträgt: <br>
l= 83,85cm = 0,8385m <br>
Das Messergebniss ist also:<br>
$ x_l= (0,84 \pm 2,04*10^{-5}) m $ <br> <br>
Bei der Zeit ist uns die Ableseungenauigkeit für die digitale Stoppuhr und die durchschnittliche Reaktionszeit eines Menschen als Messungenauigkeit bekannt <br>
Sie betragen: <br>
$ u_s=\frac{0,01s}{2 \cdot \sqrt 3} = 2,89*10^{-3} s $ und $u_r = 1,8*10^{-3} s$ <br> 
Die gesamte Messunsicherheit der Länge lässt sich durch das pythagoräisch quadratische addieren der beiden Messunsicherheiten ermitteln:
$ u_t = \sqrt{(2,89 \cdot 10^{-3}s)^2+(1,8 \cdot 10^{-3} s)^2} = 3,40 \cdot 10^{-3}s$ <br>
Die Periodendauer beträgt (wir haben 10 Perioden aufeinmal gemessen um die Messunsicherheiten zu minimieren): <br>
T= $ \frac{18,47s}{10} $ = 1,847s <br>
Das Ergebniss ist also: <br>
$ x_t = (1,85 \pm 3,40 \cdot 10^{-3} )s $ <br> <br><br>
Das Endergebniss erlangt man durch einsetzen der Messwerte in die Formel zur Bestimmung der Gravitationskonstante und durch Addition der Messunsicherheiten der Zeit und der Länge mit der Methode der Ableitungen <br> <br>
$ g= \frac{4\cdot\pi^2*0,84m}{{1,85s}^2} = 9,70\frac{m}{s^2 } $ <br> <br>
$u_{ges} = \sqrt{(\frac{4 \cdot \pi^2}{(1,847s)^2} \cdot 2,04*10^{-3} cm )^2+(\frac{-8 \cdot {\pi}^2 \cdot 0,84m}{(1,847s)^3}  \cdot 3,40 \cdot 10^{-3}s)^2} = 0,15 \frac {m}{s^2} $ <br> <br>
Das Endergebniss lautet:
    $ x_{ges} = (9,7 \pm 0,15) \frac {m}{s^2}  $

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit

Pendel = {'Länge (cm)': [10,20,30,40,50,60],
            'Periodendauer (s)': [7.13,9.47,11.09,12.82,14.22,15.37]} 

df_Pendel = pd.DataFrame(data=Pendel) # dieser Befehl DataFrame erzeugt ein Panda Dataframe
x_Pendel = df_Pendel['Periodendauer (s)'] # weise x die Werte der Spalte 'Periodendauer (s)' zu
y_Pendel = df_Pendel['Länge (cm)'] # weise y die Werte der Spalte 'Länge (cm)' zu

df_Pendel.plot.scatter('Periodendauer (s)','Länge (cm)', color='black',label='Messwerte')
plt.grid(True)
plt.legend()
plt.xlabel('Periodendauer (s)')
plt.ylabel('Länge (cm)')
plt.title("Periodendauer gegen die Länge des Pendels")
b, a,  = np.polyfit(x_Pendel,y_Pendel,deg=1)
plt.plot(x_Pendel, y_Pendel,'ko', label='Daten')
plt.plot(x_Pendel, b*x_Pendel+a,'r',label='Liniearer Fit')  
plt.show()
```
